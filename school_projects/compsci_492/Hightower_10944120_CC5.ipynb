```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from IPython.display import SVG
from graphviz import Source
```

```python
data = pd.read_csv("/content/drive/MyDrive/CustomerRetention.csv")

# Extract a 10000 by 10 matrix with attributes and
# a vector of the "target" values using the Exited attribute
data_X = data.loc[:,'CreditScore':'EstimatedSalary']
data_Y = data['Exited']
```

```python
np.random.seed(13)
indices = np.random.permutation(len(data_X))

data_X_train = data_X.iloc[indices[:-2000]]
data_Y_train = data_Y.iloc[indices[:-2000]]
data_X_test = data_X.iloc[indices[-2000:]]
data_Y_test = data_Y.iloc[indices[-2000:]]
```

#### Build k-NN with k=3,5,7,9. Output accuracy and confusion matrix. You can just build one, change value of k and run 4 times.


```python
clf = KNeighborsClassifier(n_neighbors=7)
clf.fit(data_X_train, data_Y_train)
preds = clf.predict(data_X_test)
accuracy = (preds == data_Y_test).mean()
accuracy
```

```python
confusion_matrix(data_Y_test, preds)
```

#### Build a decision tree for the data.  Output accuracy and confusion matrix.


```python
clf1 = tree.DecisionTreeClassifier()
clf1 = clf1.fit(data_X_train, data_Y_train)
preds1 = clf1.predict(data_X_test)
accuracy1 = (preds1 == data_Y_test).mean()
accuracy1
```

```python
confusion_matrix(data_Y_test, preds1)
```

#### Render a graph and determine which attribute is at the root.

```python
dot_data = tree.export_graphviz(clf1, out_file=None,feature_names=data.columns[2:-1])
graph = graphviz.Source(dot_data)
graph
```
