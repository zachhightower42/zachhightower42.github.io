```python
import numpy as np
import pandas as pd
df = pd.read_csv("./wages.csv")
df
```

```python
s = df['race']
df['race'] = s.str.lstrip('1234.')
d = df['education']
df['education'] = d.str.lstrip('1234.<')
f = df['jobclass']
df['jobclass'] = f.str.lstrip('1234.')
g = df['health_ins']
df['health_ins'] = g.str.lstrip('1234.')
df

```

```python
from sqlalchemy import String
from pandas.core.reshape.merge import string
from numpy.lib.function_base import append
from pandas.core import apply



def wage_class(value):
    if value >= 315.15:
        return "one percent"
    elif value >= 120.83:
        return "upper class"
    elif value >= 31.834:
        return "middle class"
    else:
        return "low income"

df['wage_classes'] = df['wage'].map(wage_class)
df
```

```python
health_dummies = pd.get_dummies(df['health_ins'],prefix='health_ins')
df = df.join(health_dummies)
```
