#### Basic Aggregation

```python
# let's create a numpy array for demonstration purpose
columns = ["school_year", "bonus_points", "bith_year"]
possible_school_year = list(range(6))
max_bonus_points = 20
possible_birth_year = list(range(1800, 2005))

student_info = []
for i in range(100): # lets add 100 students with random info
    school_year = rng.choice(possible_school_year)
    bonus_point = rng.choice(max_bonus_points)
    birth_year = rng.choice(possible_birth_year)
    student_info.append([school_year, bonus_point, birth_year])

student_info = np.array(student_info)

student_info
```

```python
# shape of our array
student_info.shape
```

```python
# summation, mean, median
np.sum(student_info), np.mean(student_info), np.median(student_info)
```

```python
# min, max of birth year
np.min(student_info[:, 2]), np.max(student_info[:, 2])
```

Aggregation functions take an additional argument specifying the axis along which the aggregate is computed. For example, we can find the minimum value within each column by specifying axis=0:

```python
# summation, mean on each column
np.sum(student_info, axis=0), np.mean(student_info, axis=0)
```

Similarly, we can find the aggregated value within each row:

```python
# summation, mean on each row
np.sum(student_info, axis=1)
```

```python
# 75 percentile of bonus point
np.percentile(student_info[:,1], q=75)
```

#### Basic Sorting / Argsort

```python
# basic sorting
x = np.array([2, 1, 4, 3, 5])
np.sort(x)
```
